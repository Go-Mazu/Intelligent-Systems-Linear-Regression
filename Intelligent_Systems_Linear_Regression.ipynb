```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
d = pd.read_csv('student_marks.csv')
```

```python
midavg = d['Midterm mark'].mean()
finavg = d['Final mark'].mean()
print ("The average midterm mark was " + str(midavg))
print ("The average final mark was " + str(finavg))

sv1 = d['Midterm mark'].std()
sv2 = d['Final mark'].std()
print ("The standard variation of the midterm mark was " + str(sv1))
print ("The standard variation of the final mark was " + str(sv2))

print ("The standardized midterm mark values are as follows:")
print((d['Midterm mark']-midavg)/sv1)

print ("The standardized final mark values are as follows:")
print((d['Final mark']-finavg)/sv2)
```

```python
x = ((d['Midterm mark']-midavg)/sv1)
y = ((d['Final mark']-finavg)/sv2)
plt.scatter(x, y)
plt.title('Standardized Data')
plt.xlabel("Midterm Mark")
plt.ylabel("Final Mark")
```

```python
plt.scatter(d['Midterm mark'], d['Final mark'])
x1 = d['Midterm mark']
y1 = d['Final mark']
plt.title('Non Standardized Data')
plt.xlabel('Midterm Mark')
plt.ylabel('Final Mark')
```

```python
x = ((d['Midterm mark']-midavg)/sv1)
y = ((d['Final mark']-finavg)/sv2)
plt.scatter(x, y)
plt.title('Standardized Initial Linear Regression Line with m, b = -0.5, 0')
plt.xlabel("Midterm Mark")
plt.ylabel("Final Mark")
m, b = -0.5, 0 #initial Cost/Error Function intercept and coefficient
plt.plot(x, m*x + b, color = 'r')
print("m = " + str(m) + " b = " + str(b))

scost1 = (sum(((m*x + b) - y)**2))/(len(d)) #Cost/Error Function as discussed in the lab manual
print("error = " + str(scost1))

```

```python
plt.scatter(((d['Midterm mark']-midavg)/sv1), ((d['Final mark']-finavg)/sv2))
plt.title('Standardized Data Linear Regression 100 Iterations 0.0001 Learning Rate')
plt.xlabel("Midterm Mark")
plt.ylabel("Final Mark")
m2 = -0.5
b2 = 0
alpha = 0.0001
iterations = 100
siz = len(d)
x = ((d['Midterm mark']-midavg)/sv1)
y =  ((d['Final mark']-finavg)/sv2)

Xi = min(x)
XI = max(x)

for i in range (iterations): #for loop to continuously cycle through values for for the linear regression coefficient m and intercept b to find the best line using gradient descent
    y0 = m2*x + b2 # determining predicted value
    PDm = (2/(siz)) * sum(-x * (y - y0)) #gradient partial derivative wrt to m
    PDb = (-2/(siz)) * sum(y - y0) #gradient partial derivative wrt to b
    m2 = m2 - (alpha * PDm) #updating the m and b terms by implementing the gradient derivatives in tandem with the learning rate
    b2 = b2 - (alpha * PDb)

Yo = min(y0)
Y0 = max(y0)

plt.plot([Xi, XI], [Yo, Y0], 'r') #establish range for x values for the line (run from min to max value of Midterm marks), and range of line for predicted value y0
print("m = " + str(m2) + " b = " + str(b2))

scost2 = (sum(((m2*x + b2) - y)**2))/(len(d)) #cost/error equation 
print("error = " + str(scost2))

```

```python
plt.scatter(((d['Midterm mark']-midavg)/sv1), ((d['Final mark']-finavg)/sv2))
plt.title('Standardized Data Linear Regression 2000 Iterations 0.0001 Learning Rate')
plt.xlabel("Midterm Mark")
plt.ylabel("Final Mark")
m2s = -0.5
b2s = 0
alpha = 0.0001
iterations = 2000
siz = len(d)
x = ((d['Midterm mark']-midavg)/sv1)
y =  ((d['Final mark']-finavg)/sv2)
Xi = min(x)
XI = max(x)
for i in range (iterations):
    y0s = m2s*x + b2s
    PDms = (2/(siz)) * sum(-x * (y - y0s))
    PDbs = (-2/(siz)) * sum(y - y0s)
    m2s = m2s - (alpha * PDms)
    b2s = b2s - (alpha * PDbs)
Yos = min(y0s)
Y0s = max(y0s)

plt.plot([Xi, XI], [Yos, Y0s], 'r')
print("m = " + str(m2s) + " b = " + str(b2s))

scost3 = (sum(((m2s*x + b2s) - y)**2))/(len(d))
print("error = " + str(scost3))
```

```python
plt.title('Standardized Error vs 100 Iterations')
plt.xlabel("Iterations")
plt.ylabel("Error")
x = ((d['Midterm mark']-midavg)/sv1)
y = ((d['Final mark']-finavg)/sv2)
iterations = 100
xaxis = range(iterations)
m4 = -0.5
b4 = 0
alpha = 0.0001
siz = len(d)
error = [] #declare list
def plotError(x, y, m4, b4, iterations, alpha):
  for i in range(iterations):
    y02 = m4*x + b4
    PDm2 = (2/(siz)) * sum(-x * (y - y02))
    PDb2 = (-2/(siz)) * sum(y - y02)
    m4 = m4 - (alpha * PDm2)
    b4 = b4 - (alpha * PDb2)
    errorfunction = sum((y-y02)**2)/siz #equation for error
    error.append(errorfunction) # can't use np so just keep appending to link the error data in the list
  return m4, b4, errorfunction, error

m4, b4, errorfunction, error = plotError(x, y, m4, b4, iterations, alpha)

plt.plot(xaxis, error)
print("m = " + str(m4) + " b = " + str(b4))

StanError = (sum(((m4*x + b4) - y)**2))/(len(d))
print("error = " + str(StanError))
```
